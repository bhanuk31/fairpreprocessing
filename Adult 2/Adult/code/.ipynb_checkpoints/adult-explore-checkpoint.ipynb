## Import packages

```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
from cvxpy import *
%matplotlib inline
```

## Load Data

```python
dataFolder = '../data/'

df = pd.read_csv(
    dataFolder + "adult.data",
    names=[
        "Age", "Workclass", "fnlwgt", "Education", "Education-Num", "Martial Status",
        "Occupation", "Relationship", "Race", "Sex", "Capital Gain", "Capital Loss",
        "Hours per week", "Country", "Income"],
        na_values="?")
```

Quantize age by decade.

```python
df['Age (decade)'] = df['Age'].apply(lambda x: np.floor(x/10.0)*10.0)
df['Age (decade)'] = df['Age'].apply(lambda x: np.floor(x/10.0)*10.0)
```

Limit education and age range, and adjust other variables.

```python
def group_edu(x):
    if x<=5:
        return '<6'
    elif x>=13:
        return '>12'
    else:
        return x
    
def age_cut(x):
    if x>=70:
        return '>=70'
    else:
        return x

# Limit education range
df['Education Years'] = df['Education-Num'].apply(lambda x : group_edu(x))

# Limit age range
df['Age (decade)'] = df['Age (decade)'].apply(lambda x : age_cut(x))

# Transform all that is non-white into 'minority'
df['Race'] = df['Race'].apply(lambda x: x if x== ' White' else 'Minority')

# Add binary income variable
df['Income Binary'] = df['Income'].apply(lambda x : 1 if x == " >50K" else 0)
```

Create dataframe with joint distribution.

```python
#features = ['Education Years','Age (decade)','Income','Sex','Race']
features = ['Age (decade)','Education Years','Income','Sex','Race','Income Binary']
dfJoint = df.groupby(features).size().reset_index()
dfJoint = dfJoint.rename(columns={0:'Count'})
dfJoint['Frequency'] = dfJoint['Count'].apply(lambda x : x/float(len(df)))
dfJoint.head()
```

Fraction of income above 50k in terms of age and educations.

```python
def plotHeatMap(df,title=''):
    dfIncome = (100.0*df.groupby(["Age (decade)","Education Years"])["Income Binary"].mean()).reset_index()
    dfIncomePivot = dfIncome.pivot("Age (decade)", "Education Years", "Income Binary")
    columnOrder = ["<6",6,7,8,9,10,11,12,'>12']
    dfIncomePivot = dfIncomePivot[columnOrder]
    sns.set(font_scale=1.5,font='sans-serif')
    fig = plt.figure(figsize=(10,7))
    sns.heatmap(dfIncomePivot,linewidths=2,annot=True, cbar_kws={"label": "Income percentage above 50K"},vmin=0,vmax=70)
    plt.title(title)
    
# Whole dataset
plotHeatMap(df,title='Income Percentage >50k for Whole Dataset')

# Female dataset
dfFemale = df.loc[df['Sex']==' Female',:]
plotHeatMap(dfFemale,title='Income Percentage >50k for Female Poulation')

# Male dataset
dfMale = df.loc[df['Sex']==' Male',:]
plotHeatMap(dfMale,title='Income Percentage >50k for Male Poulation')

# Minority dataset
dfMinority = df.loc[df['Race']=='Minority',:]
plotHeatMap(dfMinority,title='Income Percentage >50k for Minority Poulation')
```

# Core Optimization

First we create the optimization variables for the general setting. The form is 

\begin{align}
	\min_{p_{\Xh,\Yh|X,Y,D}}& \Delta\left(p_{X,Y},p_{\Xh,\Yh}\right) \\
	\sto~~&  J\left(p_{\Yh|D}(y|d),p_Y(y)\right)\leq \epsilon_{d,y}~\forall ~d\in \calD, y\in \{0,1\},\\
	&\Pr\left( d(x,y,\Xh,\Yh)\geq \delta^{(k)} \mid X=x,Y=y\right) \leq c_{x,y}^{(k)}~\forall~(x,y)\in \calX\times \calY,\\
	&p_{\Xh,\Yh|X,Y,D} \mbox{ is a valid distribution.}
\end{align}

We will start by letting the discriminatory variable $D$ to be gender and race, $Y$ to be income, $X$ to be education and and age. The data is taken from the "joint", reduced dataset. Probabilities will be approximated by the relative frequency in the dataset (`dfJoint`). We'll take both $J$ and $\Delta$ to be $D_{\mathrm{KL}}(\cdot \| \cdot)$, and assume $\epsilon$ and $c$ constant across all variables.

The goal is to find a matrix/dataframe $\mathbf{P}$, denoted stored in variable `dfP`, that represents the randomized mapping $p_{\Xh,\Yh|X,Y,D}$. Throughout the code below, `dfJoint` will be used as the "addressing" dataframe for $\mathbf{P}$: the (age, education), income, gender, and race combination of row $i$ of `dfJoint` will serve as the corresponding $X,D$ and $Y$ variables in row $i$ of the randomized mapping $\mathbf{P}$. The columns of `dfP`, in turn, denote the values of $\Xh,\Yh$. Analogously, the distortion matrix will be stored in ``dfD``.

```python
# Define attributes
D_features = ['Sex','Race']
Y_features = ['Income Binary']
X_features = ['Age (decade)', 'Education Years']

# Geat values for Pandas multindex
D_values = [dfJoint[feature].unique().tolist() for feature in D_features]
Y_values = [dfJoint[feature].unique().tolist() for feature in Y_features]
X_values = [dfJoint[feature].unique().tolist() for feature in X_features]


# Create multindex for mapping dataframe
DXY_features = D_features+X_features+Y_features
DXY_values = D_values+X_values+Y_values
DXY_index = pd.MultiIndex.from_product(DXY_values, names = DXY_features)

# Create multindex for distortion dataframe
XY_features = X_features+Y_features
XY_values = X_values+Y_values
XY_index = pd.MultiIndex.from_product(XY_values, names = XY_features)

# Initialize mapping dataframe
dfP = pd.DataFrame(np.zeros((len(DXY_index),len(XY_index))),index=DXY_index, columns = XY_index)

# delete entries that do not occur in dataframe (no probability estimate)
## first convert all features with non-zero prob into list of tuples
nonzero_tuples = [tuple(l) for l in dfJoint[DXY_features].values.tolist()] 

## now check if the the tuples in the dfP index are contained in it
#existing_index = [x in nonzero_tuples for x in dfP.index.tolist()]

## Drop entries with zero probability
#dfP = dfP.loc[existing_index,:]

# Initialize distortion dataframe
dfD = pd.DataFrame(np.zeros((len(XY_index),len(XY_index))),index=XY_index, columns = XY_index)
```

Great! Now we have to initialize the distortion metric. We adopt the following (somewhat arbitrary) rules:

1. Income can be lowered, but at a cost. So we will give a distortion penalty of 1 for changing a value of $Y$ from 1 to 0, and no penalty for changing from 0 to 1. We'll then add the constraint $\Pr\left( d(x,y,\Xh,\Yh)\geq 1 \mid X=x,Y=y\right) \leq \delta$ for $\delta$ appropriately small.

2. Education level cannot be decreased. For 1-12, it can be increased in at most 1 years, never exceeding 12. For >12, no distortion in education is applied. Again, mappings hat should not happen will have a cost of 1000.

3. Age can be increased or decreased in a decade without penalty. The exception is the age of 10, which cannot be distorted. The distortion penalty, again, is 1000.

These rules are implemented below, and stored in the dataframe `dfD`.


```python
# adjust education
def adjustEdu(v):
    if v=='>12':
        return 13
    elif v=='<6':
        return 5
    else:
        return v
    
def adjustAge(a):
    if a == '>=70':
        return 70.0
    else:
        return a

# distortion metric
def getDistortion(vold,vnew):
    '''
    Distortion metric.
    
    Inputs:
    *vold : dictionary of the form {attr:value} with old values
    *vnew : dictionary of the form {attr:value} with new values
    
    Output
    *d : distortion value
    '''
    
    # value that will be returned for events that should not occur
    bad_val = 1000.0
    
    
    # Adjust education years
    eOld = adjustEdu(vold['Education Years'])
    eNew = adjustEdu(vnew['Education Years'])
        
    # Education cannot be lowered or increased in more than 1 year
    if (eNew<eOld)| (eNew>eOld+1):
        return bad_val
    
    # adjust age
    aOld = adjustAge(vold['Age (decade)'])
    aNew = adjustAge(vnew['Age (decade)'])
    
    # Age cannot be increased or decreased in more than a decade
    if np.abs(aOld-aNew)>10.0:
        return bad_val
        
    # final penalty according to income
    if vold['Income Binary']>vnew['Income Binary']:
        return 1.0
    else:
        return 0.0

# create row dictionay (rows represent old values)
# this will make it easier to compute distrotion metric
rows_tuple = dfD.index.tolist()
rows_dict = [{XY_features[i]:t[i] for i in range(len(XY_features))} for t in rows_tuple]

# create columns dictionay (columns represent new values)
cols_tuple = dfD.columns.tolist()
cols_dict = [{XY_features[i]:t[i] for i in range(len(XY_features))} for t in cols_tuple]


for i in tqdm(range(dfD.shape[0])):
    old_values = rows_dict[i]
    for j in range(dfD.shape[1]):
        new_values = cols_dict[j]
        dfD.iloc[i,j] = getDistortion(old_values,new_values)
```

With the distortion matrix in hand, we can now build the optimization program. Note that we still have to convert the distortion values into the excess distortion that will be used for the optimization program. For now we assume a constant excess distortion for each constraint.

```python
# create excess distortion values: assuming constant.

# list of constraints for excess distortion

c1 = (1,.1) # value of (delta1,c1): to keep.
c2 = (999,0) # value of (delta2,c2): value that should no happen
clist = [c1,c2]

Dmatrix = dfD.as_matrix()

# Create constraint matrix list for excess distortion
# since old values index the rows, we go through the D matrix line by line, marking as 1 events where
# the threshold is violated. This will be multiplied by the probability matrix, resulting in the excess
# distortion metric

CMlist = [np.zeros(Dmatrix.shape) for i in range(len(clist))]

for x in range(len(CMlist)):
    c = clist[x][0]
    for i in range(Dmatrix.shape[0]):
        for j in range(Dmatrix.shape[1]):
            if Dmatrix[i,j] >= c:
                CMlist[x][i,j] =1.0
```

One thing is still missing: a vector of values $p_{X,Y}$ consistent with the ordering in the dataframe `dfD`. We also need masks for computing $p_{X,Y}$ from $p_{X,Y,D}$. We'll build it next.

```python
dfPxyd = pd.DataFrame(index=dfP.index,columns=['Frequency'])
index_list = [list(x) for x in dfPxyd.index.tolist()]

# find corresponding frequency value
i = 0
for comb in dfJoint[DXY_features].values.tolist():
    idx = index_list.index(comb) # get the entry corresponding to the combination
    # add marginal to list
    dfPxyd.iloc[idx,0] = dfJoint.loc[i,'Frequency']
    i+=1

    
# this is a general function that creates a mask based on overlapping,
# multilevel indices. It is not a beautiful function, but it's what we have.
def getMask(dfRef):
    # generates a mask assuming the multindex column is a subset of the multindex rows
    dfRows = pd.DataFrame(index = dfRef.index).reset_index()
    dfCols = pd.DataFrame(index = dfRef.columns).reset_index()
    target_ix = list(dfRef.columns.names)
    for i in range(dfRef.shape[0]):
        val1 = dfRows.loc[i,target_ix].tolist()
        for j in range(dfRef.shape[1]):
            val2 = dfCols.loc[j,target_ix].tolist()
            if all([x == y  for (x,y) in zip(val1,val2)]):
                dfRef.iloc[i,j] = 1.0
                
    return dfRef



# create mask that reduces Pxyd to Pxy
# so Pxyd.dot(dfMask1) = Pxy
dfMask_Pxyd_to_Pxy = pd.DataFrame(np.zeros((len(dfP),len(dfD))),index = dfP.index, columns = dfD.index)
dfMask_Pxyd_to_Pxy = getMask(dfMask_Pxyd_to_Pxy)

# compute mask that reduces Pxyd to Pyd
YD_features_index = dfJoint.groupby(Y_features+D_features)['Frequency'].sum().index
dfMask_Pxyd_to_Pyd = pd.DataFrame(np.zeros((len(dfP),len(YD_features_index))),index = dfP.index,columns = YD_features_index)
dfMask_Pxyd_to_Pyd = getMask(dfMask_Pxyd_to_Pyd)

# get  matrix for p_yd, with y varying in the columns
dfD_to_Y_address = pd.DataFrame(range(len(list(dfMask_Pxyd_to_Pyd))),index=dfMask_Pxyd_to_Pyd.columns)
dfD_to_Y_address = pd.pivot_table(dfD_to_Y_address.reset_index(),columns=D_features,index=Y_features,values=0)

# compute mask that reduces Pxyd to Py
y_index = dfD_to_Y_address.index
dfMask_Pxyd_to_Py = pd.DataFrame(np.zeros((len(dfP),len(y_index))),index = dfP.index,columns = y_index)
dfMask_Pxyd_to_Py = getMask(dfMask_Pxyd_to_Py)

# compute mask that reduces Pxy to Py
dfMask_Pxy_to_Py = pd.DataFrame(np.zeros((len(list(dfP)),len(y_index))),index = dfP.columns,columns = y_index)
dfMask_Pxy_to_Py = getMask(dfMask_Pxy_to_Py)

# compute mask that reduces Pxyd to Pd
dfMask_Pxyd_to_Pd = pd.DataFrame(np.zeros((len(dfP),dfD_to_Y_address.shape[1])),index = dfP.index,columns = dfD_to_Y_address.columns)
dfMask_Pxyd_to_Pd = getMask(dfMask_Pxyd_to_Pd)
```

## Optimization problem

Finally we are ready to instantiate the optimization in cvxpy.

```python
Pmap = Variable(dfP.shape[0],dfP.shape[1])  # main conditional map
#PmapXY = Variable(dfP.shape[1],dfP.shape[1])  # main conditional map
PXhYh = Variable(dfMask_Pxyd_to_Pxy.shape[1]) # marginal distribution of (Xh Yh)
PYhgD = Variable(dfD_to_Y_address.shape[1],dfD_to_Y_address.shape[0]) # rows represent p_(y|D)
PYhD = Variable(dfD_to_Y_address.shape[1],dfD_to_Y_address.shape[0])

# marginal distribution
dfMarginal = dfJoint.groupby(DXY_features)['Frequency'].sum()
PxydMarginal = pd.concat([dfP2,dfTest],axis=1).fillna(0)['Frequency'].values
PyMarginal = PxydMarginal.dot(dfMask_Pxyd_to_Py).T
PdMarginal = PxydMarginal.dot(dfMask_Pxyd_to_Pd).T
PxyMarginal = PxydMarginal.dot(dfMask_Pxyd_to_Pxy).T
# add constraints

# valid distribution
constraints = [sum_entries(Pmap, axis=1)==1]
constraints.append(Pmap>=0)

# definition of marginal PxhYh
constraints.append(PXhYh == sum_entries(np.diag(PxydMarginal)*Pmap,axis=0).T)

# Define the joint distribution P_{YH,D}
constraints.append(PYhD == (dfMask_Pxyd_to_Pd.values.T).dot(np.diag(PxydMarginal))*Pmap*dfMask_Pxy_to_Py.values)

# add the conditional mapping
constraints.append(PYhgD == np.diag(PdMarginal**(-1))*PYhD)

# add excess distorion constraints
for i in range(len(CMlist)):
    Pxy_xhyh = (dfMask_Pxyd_to_Pxy.values.T).dot(np.diag(PxydMarginal))*Pmap
    Cost = mul_elemwise(CMlist[i],Pxy_xhyh)
    constraints.append(Cost<=clist[i][1])

# epsilon definition for discrimination control
epsilon = 1
for d in range(dfMask_Pxyd_to_Pd.shape[1]):
    constraints.append(sum(kl_div(PYhgD[d,:].T,PyMarginal))<=epsilon)
    
# add objective
obj = Minimize(norm(PXhYh-PxyMarginal, 2))

prob = Problem(obj,constraints)
prob.solve(solver='CVXOPT',verbose=True)
```
