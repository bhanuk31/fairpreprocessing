```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
from cvxpy import *
from DTools import *
from matplotlib import gridspec
from matplotlib.patches import Rectangle
%matplotlib inline
```

# Data prep

Import the adult dataset, and name categories.

```python
dataFolder = '../data/'

df = pd.read_csv(
    dataFolder + "adult.data",
    names=[
        "Age", "Workclass", "fnlwgt", "Education", "Education-Num", "Martial Status",
        "Occupation", "Relationship", "Race", "Gender", "Capital Gain", "Capital Loss",
        "Hours per week", "Country", "Income"],
        na_values="?")
```

Group age by decade.

```python
df['Age (decade)'] = df['Age'].apply(lambda x: np.floor(x/10.0)*10.0)
df['Age (decade)'] = df['Age'].apply(lambda x: np.floor(x/10.0)*10.0)
```

Fix minority, education and age attributes.

```python
def group_edu(x):
    if x<=5:
        return '<6'
    elif x>=13:
        return '>12'
    else:
        return x
    
def age_cut(x):
    if x>=70:
        return '>=70'
    else:
        return x

# Limit education range
df['Education Years'] = df['Education-Num'].apply(lambda x : group_edu(x))

# Limit age range
df['Age (decade)'] = df['Age (decade)'].apply(lambda x : age_cut(x))

# Transform all that is non-white into 'minority'
df['Race'] = df['Race'].apply(lambda x: x if x== ' White' else 'Minority')

# Add binary income variable
df['Income Binary'] = df['Income'].apply(lambda x : 1 if x == " >50K" else 0)
```

We'll be interested only in age, education, income, gender and race.

```python
features = ['Age (decade)','Education Years','Income','Gender','Race','Income Binary']
#features = ['Education Years','Income','Sex','Race','Income Binary']
DT = DTools(df=df,features=features)
```

Set discriminatory features (`D_features`), binary response variable (`Y_features`) and decision features (`X_features`), and initialize the Discrimination Tools class).

```python
D_features = ['Gender','Race']
Y_features = ['Income Binary']
X_features = ['Age (decade)', 'Education Years']

DT.setFeatures(D=D_features,X=X_features,Y=Y_features)
```

Now we set the distortion metric. This function will receive the two dictionary of features X and Y corresponding to the new and old values, and return a distortion value. In this case the distortion function returns 1000 if education is lowered by any value, or increased by more than 1 year, or if age is increased or decreased by more than one decade. In returns a penalty of 1.0 if income is decreased. All other values are 0.

```python
class Dclass():
# adjust education
    def adjustEdu(self,v):
        if v=='>12':
            return 13
        elif v=='<6':
            return 5
        else:
            return v

    def adjustAge(self,a):
        if a == '>=70':
            return 70.0
        else:
            return a

    # distortion metric
    def getDistortion(self,vold,vnew):
        '''
        Distortion metric.

        Inputs:
        *vold : dictionary of the form {attr:value} with old values
        *vnew : dictionary of the form {attr:value} with new values

        Output
        *d : distortion value
        '''

        # value that will be returned for events that should not occur
        bad_val = 3.0


        # Adjust education years
        eOld = self.adjustEdu(vold['Education Years'])
        eNew = self.adjustEdu(vnew['Education Years'])

        # Education cannot be lowered or increased in more than 1 year
        if (eNew<eOld)| (eNew>eOld+1):
            return bad_val

        # adjust age
        aOld = self.adjustAge(vold['Age (decade)'])
        aNew = self.adjustAge(vnew['Age (decade)'])

        # Age cannot be increased or decreased in more than a decade
        if np.abs(aOld-aNew)>10.0:
            return bad_val
        
        # Penalty of 2 if age is decreased or increased
        if np.abs(aOld-aNew)>0:
            return 2.0

        # final penalty according to income
        if vold['Income Binary']>vnew['Income Binary']:
            return 1.0
        else:
            return 0.0
        
```

We set the excess distortion constraints (`c1` and `c2`) and set the distortion values. For now, we are solving the problem (to view the equation correctly you have to run the last markdown cell of this notebook with the latex preamble)

\begin{align}
	\min_{p_{\Xh,\Yh|X,Y,D}}& \sum_{x,y} \left| p_{X,Y}(x,y)- p_{\Xh,\Yh}(x,y)\right| \\
	\sto~~&  1-\epsilon\leq \frac{p_{\Yh|D}(y|d)}{p_Y(y)} \leq 1+\epsilon, \forall y\in \calY, d\in \calD\\
	&\Pr\left( d(x,y,\Xh,\Yh)\geq c_1 \mid X=x,Y=y\right) \leq \delta_1~\forall~(x,y)\in \calX\times \calY,\\
    & \Pr\left( d(x,y,\Xh,\Yh)\geq c_2\mid X=x,Y=y\right) \leq \delta_2~\forall~(x,y)\in \calX\times \calY,\\
	&p_{\Xh,\Yh|X,Y,D} \mbox{ is a valid distribution.}
\end{align}

We set `c1=.99`, `c2 = 1.99`, and `c3=999`.

```python
c1 = .99 # value of (delta1,c1): to keep.
c2 = 1.99  # value of (delta2,c2): value that should no happen
c3 = 2.99 # penalty for adjusting age
clist = [c1,c2, c3]
Dclass = Dclass()

DT.setDistortion(Dclass,clist=clist)

```

Now we optimize. We will change the value of $\epsilon$, and fix $\delta_1 = .1$, $\delta_2 = 0.05$ and $\delta_3=0$, so the events of high cost will never happen.

```python
epsilonV = np.linspace(0,.3,100)
y = np.zeros(len(epsilonV))
z = np.zeros(len(epsilonV))
dlist = [.1,0.05,0]
for i in tqdm(range(len(epsilonV))):
    try:
        DT.optimize(epsilon=epsilonV[i],dlist = dlist,verbose=False)
    except:
        y[i] = np.inf
        z[i] = np.inf
        continue
        
    y[i] = DT.optimum
    z[i] = DT.const[0]
    #print DT.optimum

#plt.plot(epsilonV,z)
```

```python
sns.set(font_scale=1.8,font='sans-serif')
plt.figure(figsize = (10,5))
ax = plt.plot(epsilonV,y,'-',linewidth=2)
plt.ylabel("Objective Value")
plt.xlabel("$\epsilon$")
plt.title("Objective vs. $\epsilon$")# for\n$\delta_1 =$"+str(dlist[0])+", $\delta_2=$"+str(dlist[1])+" and $\delta_3=$"+str(dlist[2]))
infeasible = np.where(y==np.inf)[0]
plt.axvspan(0, epsilonV[infeasible[-1]+1], color='red', alpha=0.2)
plt.xlim([epsilonV.min(),epsilonV.max()])
plt.ylim([-0.0002,y[y<np.inf].max()+0.0001])

plt.text(0.05, 0.0075, r'Infeasible', fontsize=20)
plt.tight_layout()
plt.savefig('./plots/tradeoff3.pdf')
```

# Mapping Plots

Next we make a few plots to illustrate the bias in the dataset, and how the randomized mappings overcome these biases.

First we take a look at the randomized mappings themselves.

```python
%matplotlib inline
def plotMapping(epsilon,dlist,mean_distortion =1e6,gender=' Male',race = 'Minority', income = 0, age = 30):
    #DT.optimize(epsilon=epsilon,dlist = dlist,verbose=False,mean_distortion=mean_distortion)
    dfTest = DT.dfP.reset_index()
    dfTest['Age (decade)'] = dfTest['Age (decade)'].apply(lambda x: int(x) if type(x)!=str else x)
    dfTest = dfTest.loc[(dfTest['Gender']==gender) & (dfTest['Race']== race) & (dfTest['Income Binary']== income) & (dfTest['Age (decade)']==age)].drop(['Gender','Race','Income Binary'],axis=1)
    t = [True, True]
    dfTest = dfTest.loc[:,t+(list(dfTest.sum(axis=0)>1e-1))]
    sns.set(font_scale=1.5,font='sans-serif')
    
    t = [int(x) if type(x)!= str else x for x in dfTest.columns.levels[0] ]
    dfTest.columns = dfTest.columns.set_levels(t,level=0)

    dfGroup = dfTest.groupby(['Age (decade)', 'Education Years']).mean()
    
    l = range(dfGroup.shape[0])

    v = l[-2]
    l.remove(l[-2])
    l = [v]+l
    dfGroup = dfGroup.iloc[l,:]
    l = dfGroup.columns.tolist()

    def custom_cmp(a, b):
        if a[0] > b[0]:
            return 1
        elif a[0] == b[0]:
            if a[1] != b[1]:
                if a[1] == '<6':
                    return -1
                elif b[1] =='<6':
                    return 1
                elif a[1] == '>12':
                    return 1
                elif b[1] == '>12':
                    return -1
                else:
                    return 2*int(a[1]>b[1])-1
            else:
                return 2*int(a[2]>b[2])-1
        else:
            return -1

    l.sort(custom_cmp)
    dfGroup = dfGroup[l]
    dfGroup
    
    
    sns.set_style("white")
    fig = plt.figure(figsize=(.6*dfGroup.shape[1],.5*dfGroup.shape[0]))
    ax = sns.heatmap(dfGroup,vmin=0,vmax=1.05,linewidths=0.1, linecolor='black',edgecolors='black' )
    ax.spines['right'].set_visible(True)
    ax.spines['bottom'].set_visible(True)
    ax.spines['top'].set_visible(True)
    ax.spines['left'].set_visible(True)
    if income == 0 :
        income_str = 'Low'
    else:
        income_str = 'High'
    ax.set_title('Mapping for '+str(int(age))+' Year Old '+ race+ gender+' with '+income_str +' Income')#\nfor $\epsilon=$'+str(epsilon))



 
        
epsilon = .15
dlist = [.1,.05,0]
mean_distortion = 1e6
#cmap = 'Greys'
#sns.set_style(style='dark')
DT.optimize(epsilon=epsilon,dlist = dlist,verbose=False,mean_distortion=mean_distortion)
plotMapping(epsilon,dlist,mean_distortion=mean_distortion,gender = ' Female', race =  'Minority', income = 0,age=40)

plt.savefig('./plots/map1_3.pdf',bbox_inches='tight')
plotMapping(epsilon,dlist,mean_distortion=mean_distortion,gender = ' Male', race =  ' White', income = 0,age=40)

plt.savefig('./plots/map2_3.pdf',bbox_inches='tight')
plotMapping(epsilon,dlist,mean_distortion=mean_distortion,gender = ' Female', race =  'Minority', income = 0,age=20)

plt.savefig('./plots/map3_3.pdf',bbox_inches='tight')
```

Next we randomize the dataframe. But first, we take a look at the expected marginals.

```python
DT.computeMarginals()
DT.dfPyhgD[1]/(DT.df['Income Binary'].mean())
```

```python
df2 = df.copy()
dfP2 = DT.dfP.applymap(lambda x : 0 if x<1e-8 else x)
dfP2 = dfP2.divide(dfP2.sum(axis=1),axis=0)
for idx in tqdm(df2.index):
    rowTest = df2.loc[idx,:]
    vals = rowTest[D_features+X_features+Y_features]
    draw = dfP2.loc[tuple(vals.tolist())]
    #randomly select value

    mapVal = np.random.choice(range(len(draw)),p=draw.tolist())
    draw.index[mapVal]
    df2.loc[idx,draw.index.names] = draw.index[mapVal]
#vals.tolist()
```

Great! With the dataset randomized, we now see how the income above 50k percentage of each group changed. Note that these values may change unless the numpy seed has been fixed, since we randomized the dataset sample by sample. For the plots, there is a parameter `cut`, which indicates that we do not plot the results for age-education groups with less than `cut` people (default is 30).

```python
sns.set_style("darkgrid")
sns.set(font_scale=1.8,font='sans-serif')
label_size = 30
plt.rcParams['xtick.labelsize'] = label_size 
plt.rcParams['ytick.labelsize'] = label_size 
plt.rcParams['axes.labelsize'] = label_size 

def plotHeatMapDiff(df1,df2,title='',cut = 20,ax = '',cbar=True,cbar_ax= None,xticklabels=True, yticklabels=True):
    dfIncome1 = (100*df1.groupby(["Age (decade)","Education Years"])["Income Binary"].mean()).reset_index()
    dfIncomePivot1 = dfIncome1.pivot("Age (decade)", "Education Years", "Income Binary")
    
    
    dfIncome2 = (100*df2.groupby(["Age (decade)","Education Years"])["Income Binary"].mean()).reset_index()
    dfIncomePivot2 = dfIncome2.pivot("Age (decade)", "Education Years", "Income Binary")
    
    # mask values that have count below cut -- otherwise results can be exagerated.
    dfMask = (df1.groupby(["Age (decade)","Education Years"])["Income Binary"].count()).reset_index()
    dfMaskPivot = dfMask.pivot("Age (decade)", "Education Years", "Income Binary")
    dfMaskPivot = dfMaskPivot.applymap(lambda x : np.nan if x<=cut else 0)
    
    dfIncomePivot = dfIncomePivot2-dfIncomePivot1
    dfIncomePivot = dfIncomePivot + dfMaskPivot
    
    columnOrder = ["<6",6,7,8,9,10,11,12,'>12']
    dfIncomePivot = dfIncomePivot[columnOrder]
    t = [str(int(x)) if type(x)!= str else x for x in dfIncomePivot.index ]
    dfIncomePivot.index = pd.Index(t,name=dfIncomePivot.index.name)
    
    
    sns.set(font_scale=2,font='sans-serif')
    
    if ax == '':
        fig = plt.figure(figsize=(10,7))
        sns.heatmap(dfIncomePivot,linewidths=2,annot=True, cbar=cbar,cbar_kws={"label": "Difference of income % above 50K"},vmin=-50,vmax=50,xticklabels=xticklabels, yticklabels=yticklabels)
        plt.title(title)
    else:
        if cbar_ax == None:
            sns.heatmap(dfIncomePivot,linewidths=2,annot=True,cbar=cbar, cbar_kws={"label": "Difference of income % above 50K"},vmin=-50,vmax=50,ax=ax,xticklabels=xticklabels, yticklabels=yticklabels)
            ax.set_title(title,fontsize=30)
        else:
            sns.heatmap(dfIncomePivot,linewidths=2,annot=True,cbar=cbar, cbar_ax = cbar_ax, cbar_kws={"label": "Difference of income % above 50K"},vmin=-50,vmax=50,ax=ax,xticklabels=xticklabels, yticklabels=yticklabels)
            ax.set_title(title,fontsize=30)
    
    
# Whole dataset
#plotHeatMap(df,title='Income Percentage >50k for Whole Dataset')
df3 = df.copy()
df3["Income Binary"] = df2["Income Binary"]

# Whole dataset
f, axarr = plt.subplots(1,5, sharey=False,figsize=(48,8),gridspec_kw = {'width_ratios':[4,4,4,4, .25]})

plotHeatMapDiff(df,df3,title='Difference of Income Percentage >50k for Whole Dataset',ax = axarr[0],cbar=False)

# Female dataset
dfFemale1 = df.loc[df['Gender']==' Female',:]
dfFemale2 = df3.loc[df3['Gender']==' Female',:]
plotHeatMapDiff(dfFemale1,dfFemale2,title='Female Poulation',ax = axarr[1],cbar=False)

# Male dataset
dfMale1 = df.loc[df['Gender']==' Male',:]
dfMale1 = dfMale1.loc[dfMale1['Race']==' White',:]
dfMale2 = df3.loc[df3['Gender']==' Male',:]
dfMale2 = dfMale2.loc[dfMale2['Race']==' White',:]
plotHeatMapDiff(dfMale1,dfMale2,title='Male Poulation',ax = axarr[2],cbar=False)

# Minority dataset
dfMinority1 = df.loc[df['Race']=='Minority',:]
dfMinority2 = df3.loc[df3['Race']=='Minority',:]
plotHeatMapDiff(dfMinority1,dfMinority2,title='Minority Poulation',ax = axarr[3],cbar_ax = axarr[4])
```
